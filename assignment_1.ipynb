## Problem 1--All students (5 points)
Write a Python function, called `find_max()`, to find and return the largest number in a non-empty list of numbers. E.g., if `nums = [1, 2, 3, 4, 5]`, `find_max(nums)` should return `5`. 

_Note:_ Python has a builtin `max()` function; **don't use it for this exercise!** The goal here is to practice solving basic programming problems on your own.

_Purpose:_ To give you practice solving basic programming problems with Python and working with array-like data structures.

```python
def find_max(nums):

    sorted_nums = sorted(nums)
    return sorted_nums[-1]

```

Create and test some sample lists to demonstrate that your `find_max()` function works correctly.

```python
nums = [1, 2, 3, 4, 5]
result = find_max(nums)
print(result)
```

## Problem 3--All students (10 points)

**3a**: Using the `world_population_data.csv` file, make a plot showing the population of a country of your choosing over time from 1970 to 2023. 


```python
# Your code here for 3a:
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
# If you haven't loaded the data yet, make sure to load it first

# Select data for a specific country (e.g., India)
country_data = df[df['country'] == 'China']
country_data

# Extract relevant columns for the plot
years = ['2023', '2022', '2020', '2015', '2010', '2000', '1990', '1980', '1970']
#population = country_data['2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']
selected_columns = ['2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']
population = list(country_data[selected_columns].iloc[0])


# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(years[::-1], population[::-1], marker='o', linestyle='-')
plt.title('Population of China (1970-2023)')
plt.xlabel('Year')
plt.ylabel('Population Change (1970-2023)')
plt.grid(True)
plt.show()
```

## Problem 4--Graduate Students only (10 points)

**4a**: Make a line plot showing population by continent over time from 1970 to 2023

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df
# If you haven't loaded data yet, make sure to load it first

# Extract relevant columns for the plot
#years = df['Year'].unique()
continents = df['continent'].unique()

# Create a line plot for each continent
years = ['2023', '2022', '2020', '2015', '2010', '2000', '1990', '1980', '1970']
plt.figure(figsize=(12, 8))
for continent in continents:
    continent_data = df[df['continent'] == continent]
    #population_by_year = continent_data.groupby('continent')['1970 population'].sum()
    #print(population_by_year)
    #plt.plot(population_by_year.index, population_by_year, label=continent)

    for continent, continent_data in continent_data.groupby('continent'):
        #plt.figure(figsize=(12, 8))
        #plt.figure(figsize=(8, 5))
        #population = country_data['2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']
        selected_columns = ['2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']
        population = list(continent_data[selected_columns].sum())
        plt.plot(years[::-1], population[::-1], label=continent)
plt.title('Population by Continent (1970-2023)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.legend(title='Continent', loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

**4b**: The data has a "density (km²)" column, but that is for an unknown year. Using the "area (km²)" and population data for a **continent** of your choosing, plot the density over time for **each country** in that continent. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df
# If you haven't loaded data yet, make sure to load it first

# Select data for the 'Asia' continent
asia_data = df[df['continent'] == 'Asia']

# Calculate population density for each country and each year
asia_data['Population Density 1970'] = asia_data['1970 population'] / asia_data['area (km²)']
asia_data['Population Density 1980'] = asia_data['1980 population'] / asia_data['area (km²)']
asia_data['Population Density 1990'] = asia_data['1990 population'] / asia_data['area (km²)']
asia_data['Population Density 2000'] = asia_data['2000 population'] / asia_data['area (km²)']
asia_data['Population Density 2010'] = asia_data['2010 population'] / asia_data['area (km²)']
asia_data['Population Density 2015'] = asia_data['2015 population'] / asia_data['area (km²)']
asia_data['Population Density 2020'] = asia_data['2020 population'] / asia_data['area (km²)']
asia_data['Population Density 2022'] = asia_data['2022 population'] / asia_data['area (km²)']
asia_data['Population Density 2023'] = asia_data['2023 population'] / asia_data['area (km²)']
# Create a line plot for each country
years = ['2023', '2022', '2020', '2015', '2010', '2000', '1990', '1980', '1970']
plt.figure(figsize=(12, 8))

for country, country_data in asia_data.groupby('country'):
    #print(country_data)
    #plt.plot(country_data['Year'], country_data['Population Density'], label=country)
    #population = country_data['2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']
    selected_columns = ['Population Density 2023', 'Population Density 2022', 'Population Density 2020', 'Population Density 2015', 'Population Density 2010', 'Population Density 2000', 'Population Density 1990', 'Population Density 1980', 'Population Density 1970']
    population = list(country_data[selected_columns].iloc[0])

    # Set plot labels and title
    plt.figure(figsize=(8, 5))
    plt.plot(years[::-1], population[::-1], label=country)
    
    # Set plot labels and title
    plt.title(f'Population Density Over Time - {country}')
    plt.xlabel('Year')
    plt.ylabel('Population Density (people/km²)')
    plt.legend(title='Country')
    
    # Show the plot for each country
    plt.show()
```
